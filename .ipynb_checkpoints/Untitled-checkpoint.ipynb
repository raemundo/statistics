```python
import os
import pathlib
import json
import yaml
from scrapy.selector import Selector
import datetime
```

```python
folders = ['01. Categorical Data',
 '02. Displaying and comparing quantitative data',
 '03 Summarizing quantitative data',
 '04 Modeling data distributions',
 '05 Exploring bivariate numerical data',
 '06 Study design',
 '07 Probability',
 '08 Counting, permutations, and combinations',
 '09 Random variables',
 '10 Sampling distributions',
 '11 Confidence intervals',
 '12 Significance tests (hypothesis testing)',
 '13 Inference comparing two groups or populations (AP)',
 '13 Two-sample inference for the difference between groups',
 '14 Inference for categorical data (chi-square tests)',
 '15 Advanced regression (inference and transforming)',
 '16 Analysis of variance (ANOVA)']
```

```python
parts = []
```

```python
for folder in folders:
    filtered = list(filter(lambda x: x.endswith('.ipynb'), os.listdir(f'./{folder}')))
    chapters = {'part': folder, 'chapters': list(map(lambda x: {'file': f'{folder}/{x.replace(".ipynb", "")}'} , filtered))}
    parts.append(chapters)
    #print(chapters)
```

```python
parts
```

```python
with open('_toc.yml', 'w') as f:
    yaml.dump(parts, f, default_flow_style=False, sort_keys=False)
```

```python
for folder in folders[:1]:
    for filename in os.listdir(f'./{folder}'):
        if filename.endswith('.ipynb'):
            with open(f"./{folder}/{filename}") as f:
                jn = json.load(f)
            images_cell = "".join(jn['cells'][4]['source'])
            imgs = Selector(text=images_cell).xpath(".//img/@src").extract()
            new_imgs = []
            for idx, img in enumerate(imgs):
                new_img = f"imgs/{folder[:2]}-{filename[:2]}-0{idx+1}{pathlib.Path(img).suffix}"
                md = f"![{filename.replace('.ipynb', '')[3:].strip()} fig {idx+1}](./{new_img})"
                new_imgs.append(md)
                os.rename(f"./{folder}/{img}", f"./{folder}/{new_img}")
            if imgs and new_imgs:
                jn['cells'][4]['source'] = new_imgs
                with open(f"./{folder}/{filename}", "w") as f:
                    json.dump(jn, f)
```
