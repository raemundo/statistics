```python
mu_m = 178
sigma_m = 8 
mu_w = 170
sigma_w = 6
```

```python
mu_d = mu_m - mu_w
```

```python
sigma_d_sq = sigma_m ** 2 + sigma_w ** 2
```

```python
sigma_d = np.sqrt(sigma_d_sq)
```

```python
sigma_d
```

```python
z_score_0 = (0 - mu_d) / sigma_d
```

```python
z_score_0
```

```python
stats.norm.cdf(z_score_0)
```
