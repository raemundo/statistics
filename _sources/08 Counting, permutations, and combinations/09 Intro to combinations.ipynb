```python
persons = ['A', 'B', 'C', 'D', 'E', 'F']
seats = [1, 2, 3]
```

```python
p = list(itertools.permutations(persons, len(seats)))
```

```python
p
```

```python
len(p)
```

```python
c = list(itertools.combinations(persons, len(seats)))
```

```python
c
```

```python
len(c)
```
