```python
def factorial(n):
    if n == 1:
        return 1
    else: 
        return n * factorial(n - 1)
    
def permutation_without_repetation(n, r):
    if n != r:
        return factorial(n) / factorial(n-r)
    else: # zero factorial
        return factorial(n) / factorial(1)
    
def permutation_with_repetation(n, r):
    return n ** r
```

```python
letters = list(string.ascii_lowercase)
```

```python
p = list(itertools.permutations(letters, 3))
```
