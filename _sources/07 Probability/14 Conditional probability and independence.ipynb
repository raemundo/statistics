```python
df = DataFrame({
    'On time': [167, 115, 40, 8],
    'Delayed': [3, 5, 15, 12]
}, index=['Sunny', 'Cloudy', 'Rainy', 'Snowy'])
```

```python
df['Total'] = df['On time'] + df['Delayed']
```

```python
df = df.append(df.sum().rename('Total'))
```

```python
df[-1:]
```

```python
#P(Delayed)
df[-1:]['Delayed'] / df[-1:]['Total']
```

```python
df[-2:-1]
```

```python
#P(Delayed | Snowy)
df[-2:-1]['Delayed'] / df[-2:-1]['Total']
```
