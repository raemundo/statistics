|---|---|---|
| mean | $$ \mu = \frac{ \sum_{i=1}^{N} x_{i} }{N} $$  |  $$ \bar{x} = \frac{ \sum_{i=1}^{n} x_{i} }{n} $$ |
|  variance |  $$ \sigma^{2} = \frac{ \sum_{i=1}^{N} (x_{i} - \mu)^{2} }{N} $$ |  $$ S^{2} = \frac{ \sum_{i=1}^{n} (x_{i} - \bar{x})^{2} }{n} $$ $$ S_{n-1}^{2} = \frac{ \sum_{i=1}^{n} (x_{i} - \bar{x})^{2} }{n} $$ |
