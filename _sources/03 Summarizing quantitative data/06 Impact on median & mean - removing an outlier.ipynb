```python
x = np.array([80, 90, 92, 94, 96])
```

```python
y = x[1:]
```

```python
print(f'mean of x: {np.mean(x)}')
print(f'mean of y: {np.mean(y)}')
```

```python
print(f'median of x: {np.median(x)}')
print(f'median of y: {np.median(y)}')
```
