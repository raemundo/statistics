```python
data = {'Exam': ['LSAT', 'MCAT'], 'Score': [172, 37], 'Mean': [151, 25.1], 'Standard deviation': [10, 6.4]}
```

```python
df = DataFrame(data)
```

```python
df['Z-Score'] = (df['Score'] - df['Mean']) / df['Standard deviation']
```
