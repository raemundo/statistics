```python
df = DataFrame({"#'s picks": list(itertools.product([1, 2, 3], repeat=2))})
```

```python
df['X'] = [np.mean(i) for i in df["#'s picks"]]
```

```python
df
```

```python
df['X'].plot(kind='hist',
             bins=np.arange(0.5, 4, 0.5),
             width=0.02)
```
