$$ P(X) = \frac{N!}{x!(N - x)!} \pi^{x} 1-\pi^{N-x} $$

```python
n = 6
f = 0.7
```

```python
x = range(7)
```

```python
y = [special.comb(n, k) * f ** k * (1-f)**(n - k) for k in x]
```

```python
plt.bar(x, y)
```

```python
binom_data =stats.binom.rvs(size=1000,n=6,p=0.7)
```

```python
sns.distplot(binom_data, kde=True, hist_kws={'alpha':1})
plt.xlabel('Binomial')
plt.ylabel('Frequency')
```
