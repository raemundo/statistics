```python
df = DataFrame({'mean': [10, 10], 'std. dev.': [1.5, 2]}, index=['To work', 'To home'])
```

```python
df
```

```python
mean = df['mean'].sum()
```

```python
var = (df['std. dev.']**2).sum()
```

```python
std = np.sqrt(var)
```

```python
mean, var, std
```

```python
x = 25
```

```python
z_score = (x - mean) / std
```

```python
z_score
```

```python
stats.norm.cdf(z_score)
```
