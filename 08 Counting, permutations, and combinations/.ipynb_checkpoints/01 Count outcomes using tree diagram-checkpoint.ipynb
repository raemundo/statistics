```python
engines = ['4', '6']
colors = ['Red', 'Blue', 'Green', 'White']
```

```python
tree = [i for i in itertools.product(engines, colors)]
```

```python
tree
```

```python
G=nx.Graph()

for engine, color in tree:
    G.add_edge(engine, color)
    
nx.draw_networkx(G)
plt.show()
```
