```python
ft = np.arange(1, 101, 1)
```

```python
def P(cond, total_lst):
    cond_lst = list(filter(cond, total_lst))
    return len(cond_lst) / len(total_lst), cond_lst
```

```python
def P_in_a_row(p, n):
    return p**n
```

```python
p_ft, _ = P(lambda x: x <= 75, ft)
```

```python
P_in_a_row(p_ft, 3)
```
