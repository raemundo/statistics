```python
x = np.array(20 * [1] + (50 - 20)* [0])
```

```python
xbar = x.mean()
```

```python
s = x.std(ddof=1) / np.sqrt(len(x))
```

```python
z = stats.norm.pdf(0.99) # need improvement
```

```python
xbar - z * s, xbar + z * s
```

```python
stats.norm.interval(0.99, loc=xbar, scale=s)
```
