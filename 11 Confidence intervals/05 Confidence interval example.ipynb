```python
x = np.array(142 * [1] + 108 * [0])
```

```python
sns.distplot(x)
```

```python
xbar = x.mean()
s = x.std(ddof=1)
```

```python
mu_xbar = xbar
sigma_xbar = s/np.sqrt(len(x))
```

```python
stats.norm.interval(0.99, loc=mu_xbar, scale=sigma_xbar)
```
