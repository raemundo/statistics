```python
x = np.array(57 * [0] + 43 * [1])
```

```python
sns.distplot(x)
```

```python
xbar = x.mean()
s = x.std(ddof=1)
```

```python
mu_xbar = xbar
sigma_xbar = s/np.sqrt(len(x))
```

```python
stats.norm.interval(0.95, loc=mu_xbar, scale=sigma_xbar)
```

```python
(mu_xbar - 2 * sigma_xbar, mu_xbar + 2 * sigma_xbar)
```
