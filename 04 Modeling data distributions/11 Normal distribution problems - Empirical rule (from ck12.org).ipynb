```python
mu = 9.5
sigma = 1.1
```

```python
def cal_emperical_rule(mu, sigma):
    mu = 7 * [mu]
    sigma_scaler = np.arange(-3, 4, 1) 
    sigma_scales = sigma_scaler * sigma
    emperical_rule_lst = mu + sigma_scales
    return emperical_rule_lst
```

```python
x = cal_emperical_rule(mu, sigma)
```
