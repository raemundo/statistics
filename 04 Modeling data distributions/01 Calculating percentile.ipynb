```python
x = np.array(1 * [1] + 2 * [2] + 1 * [3] + 1 * [4] + 2 * [5] + 1 * [6] + 5 * [7] + 1 * [9])
```

```python
for i in [40, 55, 70, 85]:
    percentile = np.percentile(x, i)
    if round(percentile) == 6:
        print(i, percentile)
```

```python
Series(x).plot(kind='hist')
```

```python
plt.hist(x)
plt.xlabel('Hours a day')
plt.ylabel('Drivers')
plt.show()
```
